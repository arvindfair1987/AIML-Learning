Instructions

Imagine you're working for a company that receives thousands of customer service requests via email, web forms, and social media messages. These messages often contain valuable information such as customer queries, feedback, or requests for assistance. Sometimes, the customer provides their email address within the body of the text, but this data is mixed with other text, like product names, customer IDs, or issues reported.


To improve operational efficiency, the company wants to automatically extract email addresses from all incoming data, so customer service representatives can easily contact customers for follow-ups, ensuring better customer support and faster issue resolution.


Your task is to write a function extract_email_data(text) that identifies the email addresses within the text and extracts the username and domain from them


The result should be returned as a list of tuples, where each tuple contains the following structure:


(username, domain): The username is the part before the '@' symbol, and the domain is the part after the '@' symbol, including the domain suffix (e.g., gmail.com)

Input Format


A text (str) that may contain email addresses and other random text

Output Format



A list of tuples. Each tuple contains two parts:
The username (str) extracted from the part before '@' but after whitespaces or the start of a text
The domain (str) extracted from the part after '@' but before whitespaces, punctuation, or the end of the text

Constraints


Email addresses are of the form username@domain.com
There must be whitespace before and after the username and domain (e.g., alice@gmail.com)
The domain will always contain at least one dot (.)

Example Case 1


Input


'alice@gmail.com and bob@outlook.com'


Output


[('alice', 'gmail.com'), ('bob', 'outlook.com')]


Example Case 2


Input


'Here is some text with no email addresses.'


Output 

[]

````python
import re
# Taking the input
text = input()

def extract_email_data(text):
    pattern = r'[\w\.-]+@[\w\.-]+\.\w+'
    emails = re.findall(pattern, text)

    # Split into (username, domain)
    return [(email.split("@")[0], email.split("@")[1]) for email in emails]

# Print the output
print(extract_email_data(text))
````

Instructions

The HR department wants to identify top earners in the company using data recorded as immutable tuples. For fast analytics without mutation or conversion to other types, this function helps list the names of high-salaried employees using tuple-based filtering and unpacking.


A company maintains a snapshot of its employees' details as a tuple of entries. Each record contains information about an employee, including their name, department, and salary. 

Your task is to write a function filter_high_earners(employee_data, threshold) that:


Unpacks each employee record using tuple unpacking
Returns a tuple of names of employees whose salary is strictly greater than the provided threshold



Input Format

employee_data: A tuple of employee records, where each record is a tuple:
employee_name (str)
department (str)
salary (int or float)
threshold: A numeric value (int) used to filter employees

Output Format


A tuple of employee names (str) who earn more than the threshold
If no employees meet the criteria, return an empty tuple

Constraints


Names are unique strings
Salary values are positive or zero

Example Case 1


Inputs


(('Alice', 'HR', 50000),('Bob', 'Engineering', 80000),('Charlie', 'Marketing', 60000))

55000


Output


('Bob', 'Charlie')


Example Case 2


Inputs


(("Diana", "Sales", 40000), ("Eli", "Support", 42000))

50000


Output


()

```python
# Import literal_eval library to safely evaluate string input as a Python literal
from ast import literal_eval

# Taking the input
employee_data = (('Alice', 'HR', 50000),('Bob', 'Engineering', 80000),('Charlie', 'Marketing', 60000))  
threshold = 55000             

def filter_high_earners(employee_data, threshold):
    return tuple(name for name, dept, sal in employee_data if sal >= threshold)


# Print the output
print(filter_high_earners(employee_data, threshold))
```

Instructions

A retail store maintains a tuple of items, where each item is represented as a tuple of the form (item_name, price).


Your task is to write a function get_most_expensive_item(store_items) that:


Returns the name of the most expensive item
If the tuple is empty, the function must alert the user that it is so
If there is a tie in prices for the most expensive item, return the name of the first item appearing in the input tuple

This function simulates a price check in a retail storeâ€™s billing system. It helps quickly identify the highest-priced item in the current stock, or acknowledge when the store has no items.


Input Format


A tuple of items, where each item is itself a tuple:
The first element is the item name (str)
The second element is the price (int or float)

Output Format



Either the name (str) of the most expensive item OR
'No items available' if the input tuple is empty
If there is a tie in prices for the most expensive item, return the name of the first item appearing in the input tuple

Constraints


Prices can be assumed to be non-negative
All item names are unique

Example Case 1

Input
(('Pen', 10), ('Notebook', 30), ('Bag', 50))


Output
Bag


Example case 2

Input
()


Output

No items available

```python
# Import literal_eval library to safely evaluate string input as a Python literal
from ast import literal_eval

# Taking the input
items_list = [(('Pen', 10), ('Notebook', 30), ('Bag', 50)), 
         (('Laptop', 700), ('Phone', 999.99), ('Mouse', 25)),
         (),(('A', 0), ('B', 0.01), ('C', 0.001))]

def find_most_expensive(items):
    if len(items) == 0:
        return 'No items available'
    return max(items, key=lambda x: x[1])[0]


# Print the output
for items in items_list:
    print(find_most_expensive(items))
```

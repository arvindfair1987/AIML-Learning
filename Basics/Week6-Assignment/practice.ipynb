```python
def quntiles(data, q):
    sorted_data = sorted(data)
    n = len(sorted_data)

    quantile_value = (q - 1) * [0]
    for k in range(1,q):
        pos = k/q * (n-1)
        lower = int(pos)
        upper = lower + 1 
        fraction = pos - lower
        quantile_value[k-1] = sorted_data[lower] + fraction * (sorted_data[upper] - sorted_data[lower])

    return quantile_value
```

```python
print(quntiles([15,18,21,50,67,100],2))
```

```python
def comupte_mad(data):
    mean = sum(data)/len(data)
    mad = sum(abs(x - mean) for x in data) / len(data)
    return mad
```

```python

def compute_variance(data):
    mean = sum(data)/len(data)
    var = sum ( (x - mean) ** 2   for x in data) / len(data)
    return var
```

```python
import math

def compute_sd(data):
    var = compute_variance(data)
    return math.sqrt(var)
```

```python
def compute_covariance(x,y):
    n = len(x)
    mean_x = sum(x)/len(x)
    mean_y = sum(y)/len(y)
    return sum( (x[i] -mean_x)*(y[i]-mean_y) for i in range(n) )/n
```

```python
compute_covariance([2, 4, 6, 8, 10], [1, 3, 2, 5, 7])
```

```python
compute_covariance([2, 4, 6, 8, 10], [1, 3, 2, 5, 7][::-1])
```

```python
def compute_variance(data):
    mean = sum(data)/len(data)
    return sum((x-mean)** 2 for x in data) / len(data) 

```

```python
def compute_correlation(x,y):
    cov = compute_covariance(x,y)
    sd_x = compute_covariance(x,x) ** 0.5
    sd_y = compute_covariance(y,y) ** 0.5

    return cov/(sd_x * sd_y)
    
```

```python
compute_correlation([2, 4, 6, 8, 10], [1, 3, 2, 5, 7])
```

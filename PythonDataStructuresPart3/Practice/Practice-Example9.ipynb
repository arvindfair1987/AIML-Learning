Instructions

In image processing, a common task is to convert a coloured image into greyscale. Each pixel in an image is represented by three values (red, green, and blue channels), where each value is in the range [0,255]. To convert an image to greyscale, a typical approach is to calculate the average of the red, green, and blue values of each pixel.


For instance, for each pixel, the greyscale value is calculated as: 


greyscale = (red + green + blue)/3 



Write a Python function convert_to_greyscale(image) that takes an image matrix as input, where the image is represented by a list of lists, and each element in the matrix is a list containing three integers representing the red, green, and blue values of a pixel. The function should return a new matrix, where each element is the greyscale value of the pixel calculated as the average of the red, green, and blue values.



Input format


image: A list of lists, where each inner list contains three values (int) representing the red, green, and blue values of a pixel of the image; there is no restriction on the length of the outer list

Output format


If any red, green, and blue values are not in the range [0,255], return a string 'Invalid RGB values present'
Otherwise, return a list where each element (int) represents the greyscale value (average of the red, green, and blue values, rounded down) of the corresponding pixel in the original image

Constraints


The elements within the input list must be lists three elements long

Example case 1


Input


[[255, 0, 0], [0, 255, 0], [0, 0, 255]]


Output

[85, 85, 85]


Example case 2


Input


[[-1, 0, 255], [0, 0, 255]]


Output


Invalid RGB values present

```python
# Import literal_eval library to safely evaluate string input as a Python literal 
from ast import literal_eval

# Taking the input
image_list = [[[255, 0, 0], [0, 255, 0], [0, 0, 255]],
              [[-1, 0, 255], [0, 0, 255]],
              [[255, 255, 255], [0, 0, 0], [128, 128, 128], [225, 0, 112]],
              [[100, 150, 200], [50, 75, 100]],
              [[255, 0, 0], [255, 255, 255], [0, 0, 0]]]

def convert_to_greyscale(image):
    result = [sum(rgb)//3 if min(rgb)>=0 and max(rgb)<=255 else "Invalid RGB values present" for rgb in image]
    
    if "Invalid RGB values present" in result:
        return "Invalid RGB values present"
    
    return result


# Print the output
for image in image_list:
    print(convert_to_greyscale(image))
```

````python
#Vector Addition

def vector_add(v1,v2):
    return tuple(a+b for a,b in zip(v1,v2))

#vector Substract

def vector_substract(v1,v2):
    return tuple(a-b for a,b in zip(v1,v2))

#dot product

def dot_product(v1,v2):
    if len(v1) != len(v2):
        raise ValueError("Vectors must be the same length")
    return sum(tuple(a*b for a,b in zip(v1,v2)))

#magnitude of vector 

def magnitude(vector): 
    return sum(component**2 for component in vector) ** 0.5

def scale(v1, k):
    return tuple(component*k for component in v1)

import math 

def angle_between(v1,v2):
    cos_theta = dot_product(v1,v2)/(magnitude(v1) * magnitude(v2))
    cos_theta = max(-1.0, min(1.0,cos_theta)) 
    return math.degrees(math.acos(cos_theta))

def cosine_similarity(v1,v2):
    cos_theta = dot_product(v1,v2)/(magnitude(v1) * magnitude(v2))
    return  cos_theta

def project(v1,v2):
    if dot_product(v1,v2) ==0:
        return 0
    k = dot_product(v1,v2)/dot_product(v1,v2)
    return scale(v2,k)

def norm1(v1,v2):
    return sum(vector_substract(v1,v2))

def norm2(v1,v2): # for Similarities 
    return magnitude(vector_substract(v1,v2))

def normp(v1,p):
    return sum(x ** p for x in v1) ** (1/p)

def vectors_2_normp(v1,v2,p):
    diff = vector_substract(v1,v2)
    return normp(diff,p)


def liner_combination(vectors, coeffienents):
    no_dimensions = len(vectors[0])
    for vector in vectors:
        if len(vector) != no_dimensions:
            print("Dimensions are not equal")
            break
    
    result = [0]*no_dimensions
    for i in range(no_dimensions):
        for j,vector in enumerate(vectors):
            result[i] += coeffienents[j] * vector[i]

    return tuple(result)

def linear_combination2(vectors, coeffienents):
    if any(len(v) != len(vectors[0]) for v in vectors):
         raise ValueError("Vectors must have the same dimension")
    return tuple( 
        sum(c * v[i] for c,v in zip(coeffienents,vectors) for i in range(len(vectors[0])))
    )


````

````python
#Matrics Operations
def matrics_addition(A,B):
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

def matrics_substraction(A,B):
    return [ [A[i][j] - B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

def matrics_scalar(A,k):
    return [[k*A[i][j] for j in range(len(A[0]))] for i in range(len(A))]
 
def matrics_transpose(A):
    return [[A[j][i] for j in range(len(A[0]))] for i in range(len(A))]
````

## ðŸ§  What Matrix Multiplication Means (EXAM GOLD)

If

* **A** is of size ( m \times n )
* **B** is of size ( n \times p )

Then

* **Result** is of size ( m \times p )

ðŸ“Œ **Rule:**

> Number of columns in A = Number of rows in B

---

## ðŸ§© Break the Code Line by Line

### 1ï¸âƒ£ Outer loop â€“ Rows of A

```python
for i in range(len(A))
```

* `i` â†’ row index of **A**
* `len(A)` â†’ number of rows in A
  âž¡ï¸ Each `i` creates **one row** of the result matrix

---

### 2ï¸âƒ£ Middle loop â€“ Columns of B

```python
for j in range(len(B[0]))
```

* `j` â†’ column index of **B**
* `len(B[0])` â†’ number of columns in B
  âž¡ï¸ Each `j` creates **one element** in the result row

---

### 3ï¸âƒ£ Inner sum â€“ Dot Product (MOST IMPORTANT)

```python
sum(A[i][k] * B[k][j] for k in range(len(B)))
```

This is the **dot product** of:

* Row `i` of A
* Column `j` of B

#### Why `k`?

* `k` runs over the **shared dimension**
* `len(B)` = number of rows in B = number of columns in A

ðŸ“Œ Formula:
[
\text{Result}[i][j] =
\sum_{k} A[i][k] \times B[k][j]
]

## ðŸ“ Exam-Ready One-Liner Answer

> This list comprehension performs matrix multiplication by iterating over rows of matrix A and columns of matrix B, and computing each element as the dot product of the corresponding row and column.



````python
#len(A) -- Rows in A
#len(B[0]) -- Columns in B
#len(A[0]) -- columns in A 
#len(B) -- Rows in B
# For Matrics Muliplcation columns in A  should be equal to Rows in B i.e len(A[0]) == len(B) that we took as k
# then took dot product i.e sum A[i][k] * B[k][j]

def matrics_multiplication(A,B):
    return [[sum(A[i][k] * B[k][j] for k in range(len(A[0]))) for j in range(len(B[0]))] for i in range(len(A))]

#or 

def matrics_multiplication2(A,B):
    return [[sum(A[i][k] * B[k][j] for k in range(len(B))) for j in range(len(B[0]))] for i in range(len(A))]

A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

print(matrics_multiplication(A,B))
print(matrics_multiplication2(A,B))
````

````python
#Determinent of 2x2 

def determinent(A):
    return A[0][0]*A[1][1] - A[0][1] * A[1][0]

# inverse of 2X2

def inverse(A):
    det = determinent(A)
    a = A[0][0]
    b = A[0][1]
    c = A[1][0]
    d = A[1][1] 
    
    return [[d/det, -b/det],
             [-c/det, a/det] ]  

A = [[5, 6],
     [7, 8]]

print(determinent(A))
print(inverse(A))
print(matrics_multiplication(A,inverse(A)))
````
